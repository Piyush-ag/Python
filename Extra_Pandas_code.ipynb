```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Emily'],
    'Age': [25, 30, 35, 40, 45, 45],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Female'],
    'Salary': [50000, 60000, 72000, 730000, 90000, 90000],
    'Department': ['Marketing', 'Sales', 'Engineering', 'Engineering', 'Marketing', 'Marketing'],
    'Birthdate': ['1997-01-15', '1992-06-02', '1987-12-24', '1982-08-08', '1977-04-16', '1977-04-16']
}

df = pd.DataFrame(data)
df
```

```python
duplicates = df.duplicated()  # identify duplicate rows
duplicates
```

```python
df.drop_duplicates(inplace=True)  # remove duplicate rows
df
```

```python
df.values
```

```python
df.axes
```

```python
df.dtypes
```

```python
df.set_index('Name', inplace=True)  # set 'Name' as the index of the DataFrame
df
```

```python
df['Salary'] = df['Salary'].astype(float)  # convert 'Salary' to a float data type
df.dtypes
```

```python
df['Birthdate'] = pd.to_datetime(df['Birthdate'])
df.dtypes
```

```python
df['Birthdate'].dt.month # day, month, ...
```

```python
df['Salary'].apply(lambda x: x * 1.1)  # increase each 'Salary' value by 10%
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})  # convert 'Gender' to a binary data type
df
```

```python
df.query('Age > 30 and Salary > 70000')  # filter the DataFrame to include only rows with Age > 30 and Salary > 70000
```

```python
df1 = df.copy()

df1['Diff'] = df1['Salary'].diff()  # calculate the difference between consecutive rows in the 'Salary' column
df1['Pct_Change'] = df1['Salary'].pct_change()  # calculate the percentage change between consecutive rows in the 'Salary' column
df1['Shifted'] = df1['Salary'].shift(1)
df1['Moving_Average'] = df1['Salary'].rolling(window=3).mean()
df1['Cumulative_Sum'] = df1['Salary'].expanding().sum()
df1['Mean_Salary'] = df1.groupby('Department')['Salary'].transform('mean')

df1
```
