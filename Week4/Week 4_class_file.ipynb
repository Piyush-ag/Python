#### concat function

```python
import random
```

```python
nums = np.linspace(0,6,7)
norm_nums = np.random.normal(loc=20, scale=3, size=(8,3))
unif_nums = np.random.random(size = (10,2))
```

```python
df1 =  pd.DataFrame(nums, columns = ['col1'])
df2 =  pd.DataFrame(norm_nums, columns = ['col1','col2','col3'])
df3 =  pd.DataFrame(unif_nums, columns = ['col1','col2'])

print(df1, "\n")
print(df2, "\n")
print(df3)
```

```python
pd.concat([df1,df2,df3], axis=1)
```

```python
pd.concat([df1,df2,df3])
```
