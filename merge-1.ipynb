```python
dataTop = pd.read_csv('Fuel_Level_Part_1-1.csv')
dataTop.rename(columns={"Tank ID":"tank_id", "Fuel Level":"fuel_level", "Time stamp": "time_stamp"}, inplace=True)
dataBottom = pd.read_csv('Fuel_Level_Part_2-1.csv')
dataBottom.rename(columns={"Tank ID":"tank_id", "Fuel_Level":"fuel_level", "Timestamp": "time_stamp"}, inplace=True)

# Display the first few rows
print(dataTop.head())

print(dataBottom.head())

print(dataTop.dtypes)


```

```python
dataAll = pd.concat([dataTop, dataBottom]).drop_duplicates().reset_index(drop=True)
print(dataAll.head())
```

```python
import pandas as pd

# Read the CSV files into DataFrames
tanks = pd.read_csv('Tanks-1.csv')
locations = pd.read_csv('Locations-1.csv')

# Rename columns for consistency
tanks.rename(columns={
    'Tank ID': 'tank_id',
    'Tank Location': 'tank_location',
    'Tank Number': 'tank_number',
    'Tank Type': 'tank_type',
    'Tank Capacity': 'tank_capacity'
}, inplace=True)

locations.rename(columns={
    'Gas Station Location': 'gas_station_location',
    'Gas Station Name': 'gas_station_name',
    'Gas Station Address': 'gas_station_address',
    'Gas Station Latitude': 'gas_station_latitude',
    'Gas Station Longitude': 'gas_station_longitude'
}, inplace=True)

# Display the first few rows to verify changes
print(tanks.head())
print(locations.head())
```

```python
tanks_and_locations = tanks.merge(locations, left_on ="tank_location",right_on ="gas_station_location", how="inner")
tanks_and_locations.drop(columns="tank_location", inplace=True)
tanks_and_locations
```

```python
tanks_and_locations = tanks_and_locations.merge(dataAll, left_on ="tank_id",right_on ="tank_id", how="inner")
tanks_and_locations
```
