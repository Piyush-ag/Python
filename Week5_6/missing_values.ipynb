### Missing Values

```python
import numpy as np
import pandas as pd
import missingno as msno
```

```python
dirty_df = pd.read_csv("dirty_house_data-1.csv")
```

```python
missing_values_char = ["-", "--", "?", "??", "???", "na", "Na", "nA", "NA", "n/a"]
dirty_df = pd.read_csv("dirty_house_data-1.csv", na_values=missing_values_char)
dirty_df
```

```python
dirty_df.tail()
```

#### Basic missing data visualization

```python
msno.matrix(dirty_df)
```

```python
msno.bar(dirty_df)
```

```python
msno.heatmap(dirty_df)
```

#### Dropping missing values

```python
dirty_df_2 = dirty_df.drop(columns=["building_area"], index=[3, 19])
dirty_df_2
```

```python
dirty_df.dropna(thresh=9, axis=1)
```

#### panda's Mathematical operations ignore NA values

```python
dirty_df.price.mean()
```

#### Filling missing values

```python
dirty_df.price
```

```python
dirty_df.price.fillna(0)
```

```python
price_mean = dirty_df.price.mean()
price_median = dirty_df.price.median()
dirty_df.price.fillna(price_median)
```

```python
dirty_df.price
```

```python
dirty_df.price.fillna(method="ffill")
```

```python
dirty_df.price.fillna(method="ffill").fillna(method="bfill")
```

```python
dirty_df.price.interpolate()
```
